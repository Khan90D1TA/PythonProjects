```python
import matplotlib.pyplot as plt
import numpy as np

from tensorflow.keras import models, datasets, layers, optimizers
```

```python
# https://keras.io/api/datasets/fashion_mnist/

(X_train_full, y_train_full), (X_test, y_test) = datasets.fashion_mnist.load_data()
```

```python
X_train_full.shape
```

```python
X_test.shape
```

The Fashion MNIST dataset contains 70000 images. 60000 images are for training and 10000 are for testing. There are 10 different classes.

```python
X_train, X_valid = X_train_full[:50000], X_train_full[50000:]
y_train, y_valid = y_train_full[:50000], y_train_full[50000:]
```

```python
X_train.shape
```

```python
X_valid.shape
```

```python
X_train[0]  # features of index 0
```

```python
plt.imshow(X_train[0], cmap="gray")
print(y_train[0]);
```

```python
names = ['T-shirt/top', 'Trouser', 'Pullover','Dress', 'Coat',
         'Sandal', 'Shirt','Sneaker', 'Bag', 'Ankle boot']
plt.figure(figsize=(8, 8))

for i in range(25):  #25 images for predictions
  plt.subplot(5, 5, i+1)
  plt.imshow(X_train[i], cmap="gray")
  plt.xlabel(names[y_train[i]]) # identify each image by adding name
  plt.xticks([])
  plt.yticks([]);
```

```python
y_train  #50000 images
```

```python
X_train = X_train/255   #scaling  
X_valid = X_valid/255   #min max scaling 
X_test = X_test/255
```

We divide by 255 to rescale our pixel values. It is better to have the range of the values from 0 to 1 than from 0 to 255. 

```python
# Use a flatten input layer and output layer

ann1 = models.Sequential([
                        layers.Flatten(input_shape=(28, 28)), # reshapes images. flatten into a row 
                        layers.Dense(10, activation='softmax'),   
])             #

ann1.compile(optimizer='adam',  #optimizers from keras,  https://keras.io/api/optimizers/
             loss='sparse_categorical_crossentropy',  #calculates loss within our model.  
             metrics=['accuracy'])

```

```python
ann1.summary() #number of weight 784*10
#number of bias 10 
```

```python
ann1.layers[1].get_weights()[1]
```

```python
ann1.fit(X_train, y_train, epochs=5) #fit method on training data
# five iteration   
#loss function is decreasing at each epoch. 
```

```python
info = ann1.fit(X_train, y_train, epochs=15, validation_data=(X_valid, y_valid))
#we choose 15 epochs
```

```python
acc = info.history['accuracy']
val_acc = info.history['val_accuracy']

loss = info.history['loss']
val_loss = info.history['val_loss']

plt.figure(figsize=(8, 8))

plt.subplot(2, 1, 1)
plt.plot(acc, label='training data')
plt.plot(val_acc, label='validation data')
plt.ylim([0.90,.75])
plt.xlabel('epoch')
plt.ylabel('accuracy')
plt.legend()

plt.subplot(2, 1, 2)
plt.plot(loss, label='training data')
plt.plot(val_loss, label='validation data')
plt.ylim([0.3,0.7])
plt.xlabel('epoch')
plt.ylabel('loss')
plt.legend(); 
#loss in training is going down—always.
#loss in validation stays between 0.43 to 0.45—approximately.
```

```python
ann1.evaluate(X_test, y_test)    # 
```

```python
y_pred = ann1.predict(X_test)
y_pred[0] #first data observation 
#output has 10 probabilites between 0 and 1
#the predicted output will be the largest probrabilty 
# which is  7.88686514e-01 or the 10th probability/index 9
```

```python
# getting the index with the highest probability
np.argmax(y_pred[0]) 
```

```python
plt.imshow(X_test[0], cmap="gray")
print(y_test[0]); #actual label
print(names[y_test[0]])
```

```python
# Using hidden layers

ann2 = models.Sequential([
    layers.Flatten(input_shape=(28, 28)),
    layers.Dense(128, activation='relu'),
    layers.Dense(10, activation = 'softmax')
])

ann2.compile(optimizer='adam', 
             loss='sparse_categorical_crossentropy',
             metrics=['accuracy'])

ann2.fit(X_train, y_train, epochs=7)
```

```python
ann2.evaluate(X_valid, y_valid)
#highest accuracy compared to other models
```

```python
y_pred = ann2.predict(X_test)
y_pred[0] 
# highest probability: 9.7628069e-01
```

```python
np.argmax(y_pred[0]) 
plt.imshow(X_test[0], cmap="gray")
print(y_test[0]); #actual label
print(names[y_test[0]])
```

```python
# Use dropout and batch normalization   
# reduce overfitting
# 

from keras.layers import Dropout, BatchNormalization

ann3 = models.Sequential([
    layers.Flatten(input_shape=(28, 28)),  #dropout layer
    
    layers.Dense(128, activation='relu'),
    layers.BatchNormalization(),
    layers.Dropout(0.2),
    layers.Dense(10, activation='softmax')
])

ann3.compile(optimizer='adam',
             loss='sparse_categorical_crossentropy',
             metrics=['accuracy'])

ann3.fit(X_train, y_train, epochs=10)
```

```python
ann3.evaluate(X_valid, y_valid)    
```

```python
ann3.summary()
```

```python
#ann3.layers[1].get_weights()[1]
```

```python
y_pred = ann3.predict(X_test)
y_pred[3] 
# highest probability: 9.9996972e-01
```
